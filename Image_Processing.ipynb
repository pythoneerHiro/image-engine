```python
import numpy as np
import requests
from io import BytesIO

import typing 

from PIL import Image

def retreiveImage(url: str) -> Image:
  import requests
  from io import BytesIO

  response = requests.get(url)
  img = Image.open(BytesIO(response.content))
  print(img.size,img.mode,img.format)
  return img

url = "https://medias.spotern.com/spots/w640/57/57906-1532336916.webp"
url = "https://static.wikia.nocookie.net/big-hero-6-fanon/images/0/0f/Hiro.jpg/revision/latest?cb=20180511180437"
img = retreiveImage(url)
img
```

```python
from icecream import ic
from typing import Tuple


def genTilesFromImage(img:Image.Image,grid:Tuple[int,int]=(3,3))->List[np.ndarray]:
  im = np.array(img)
  M = im.shape[0]//grid[0]
  N = im.shape[1]//grid[1]
  ic(im.shape,M,N,grid) 
  tiles = [im[x:x+M,y:y+N] for x in range(0,im.shape[0],M) for y in range(0,im.shape[1],N)]
  ic(len(tiles))
  return tiles
```

```python
tiles = genTilesFromImage(img,(4,4))
```
